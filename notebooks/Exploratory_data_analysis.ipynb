### Handling Missing Values
- Drop column with high percentage missing values > 50%.
  * Unnamed column (100% missing)

```python
# Drop colmuns with high missing values 
df_cleaned = p.drop_high_missing_columns(df)
```

This data are scrapped from world bank, so some countries have total or some of missing values in the year column because of no record.

```python
# Display all columns
pd.set_option('display.max_columns', None)

# Identify numerical columns
numerical_columns = df_cleaned.select_dtypes(include=['number']).columns

# Calculate the percentage of missing values for each row, considering only numerical columns
df_cleaned['missing_percentage'] = df_cleaned[numerical_columns].isna().sum(axis=1) * 100 / len(numerical_columns)

# Filter rows with any missing values in numerical columns
filtered_rows = df_cleaned[df_cleaned['missing_percentage'] > 0]

# Display the 'Country Name' and 'missing_percentage' columns
print("\nRows with missing values in numerical columns and their percentages:")
print(filtered_rows[['Country Name', 'missing_percentage']])
```

```python
# Drop rows with high missing values
df_cleaned = p.drop_high_missing_rows(df_cleaned)
```

After cleaning rows with 100 % missing values in numerical columns we have to fill the other with 0.

```python
df_cleaned = df_cleaned.fillna(0)
```

```python
# Display all columns
pd.set_option('display.max_columns', None)

# Identify numerical columns
numerical_columns = df_cleaned.select_dtypes(include=['number']).columns

# Calculate the percentage of missing values for each row, considering only numerical columns
df_cleaned['missing_percentage'] = df_cleaned[numerical_columns].isna().sum(axis=1) * 100 / len(numerical_columns)

# Filter rows with any missing values in numerical columns
filtered_rows = df_cleaned[df_cleaned['missing_percentage'] > 0]

# Display the 'Country Name' and 'missing_percentage' columns
print("\nRows with missing values in numerical columns and their percentages:")
print(filtered_rows[['Country Name', 'missing_percentage']])
```

#### Save Cleaned data to csv file

```python
# Drop the 'missing_percentage' column
df_cleaned = df_cleaned.drop(columns=['missing_percentage'])

# Save cleaned data
df_cleaned.to_csv('../data/population_data.csv')
```
